```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import seaborn as sns
```

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/JhonatanUscca/Estructuras-de-datos-avanzadas-B/main/laboratorio01/Distancias.csv", sep=';')
df2.head()
```

```python
plt.hist(df2['Distancia_10'], bins=15, color= "yellow", ec="black")
plt.xlabel('Distancia')
plt.ylabel('Frecuencia')
plt.title('Distancia entre puntos con dimension 10')
plt.show()
```
